```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

tf.random.set_seed(42)

import random
random.seed(42)
import numpy as np
np.random.seed(42)
```

```python
from pandas_ods_reader import read_ods

df = read_ods('Biogas data.ods', 'Sheet1')
```

```python
num = df._get_numeric_data()
```

```python
x = num.iloc[:, :]

y = num.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler_vae = StandardScaler()

x_full_scaled = scaler_vae.fit_transform(x)
```

```python
print(x_full_scaled)
```

# Variational Autoencoder

```python
vae_mft = keras.models.load_model('vae.tf')
history = vae_mft.fit(x_full_scaled, x_full_scaled, epochs=500)
 
```

```python
sample_size = x_full_scaled.shape[:][1]

sample_number = 250

x_simulated = np.empty((sample_number,sample_size))
 
for i in range(sample_number):
    x_simulated[i] = vae_mft.predict(x_full_scaled[[random.randrange(x_full_scaled.shape[0])]])

    
'''import matplotlib.pyplot as plt

plt.figure(figsize=(20,5))
plt.plot(x_full_scaled.flatten(), label='x real data')
plt.plot(x_simulated.flatten(), label='x simulated data')
plt.legend(loc='upper left')
plt.show()'''
```

```python
x_full = np.concatenate([x_full_scaled, x_simulated])
```

```python
print(x_full.shape)
```

```python
rev_x = scaler_vae.inverse_transform(x_full)
print(rev_x)
```

```python
x = rev_x[:, 0:-1]

y = rev_x[:, -1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
scaler_x = StandardScaler()
x_train_scaled = scaler_x.fit_transform(x_train)
x_valid_scaled = scaler_x.transform(x_valid)
x_test = scaler_x.transform(x_test)
```

```python
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1))
y_valid_scaled = scaler_y.transform(y_valid.reshape(-1,1))
```

# Training

```python
model = keras.models.Sequential([
    keras.layers.Dense(12, activation="relu", input_shape=x_train_scaled.shape[1:]),

    keras.layers.Dense(44, activation="relu"),

    keras.layers.Dense(1)])
```

```python
def r2_score(y_true, y_pred):
    SS_res = keras.backend.sum(keras.backend.square(y_true - y_pred))
    SS_tot = keras.backend.sum(keras.backend.square(y_true - keras.backend.mean(y_true)))
    return (1 - SS_res / (SS_tot + keras.backend.epsilon()))


optimizer = tf.keras.optimizers.SGD(learning_rate=0.01,
                                    momentum=0.9,
                                    nesterov=True)


model.compile(loss="mse", optimizer=optimizer, metrics=[tf.keras.metrics.RootMeanSquaredError(), r2_score])
```

```python
history = model.fit(x_train_scaled, y_train_scaled, epochs = 150,
                    validation_data=(x_valid_scaled, y_valid_scaled))
```

```python
x_new = x_test # pretend these are new instances
y_pred = model.predict(x_new)
```

```python
print(scaler_y.inverse_transform(y_pred[::10]), y_test[::10])
```

```python
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error
y_pred = scaler_y.inverse_transform(y_pred)
y_valid = scaler_y.inverse_transform(y_valid_scaled)
y_train = scaler_y.inverse_transform(y_train_scaled)
x_valid = scaler_y.inverse_transform(model.predict(x_valid_scaled))
x_train = scaler_y.inverse_transform(model.predict(x_train_scaled))
```

```python
print(r2_score(y_pred, y_test), np.sqrt(mse(y_pred, y_test)))
print(r2_score(x_valid, y_valid), np.sqrt(mse(x_valid, y_valid)))
print(r2_score(x_train, y_train), np.sqrt(mse(x_train, y_train)))
```

```python
from matplotlib import pyplot

pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='valid')
pyplot.legend()
pyplot.show()
```
