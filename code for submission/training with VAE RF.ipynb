```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

tf.random.set_seed(42)

import random
random.seed(42)
import numpy as np
np.random.seed(42)
```

```python
from pandas_ods_reader import read_ods

df = read_ods('Biogas data.ods', 'plant1')
```

```python
num = df._get_numeric_data()
```

```python
x = num.iloc[:, :]

y = num.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler_vae = StandardScaler()

x_full_scaled = scaler_vae.fit_transform(x)
```

```python
print(x_full_scaled)
```

# Variational Autoencoder

```python
vae_mft = keras.models.load_model('vae.tf')
history = vae_mft.fit(x_full_scaled, x_full_scaled, epochs=500)
 
```

```python
sample_size = x_full_scaled.shape[:][1]

sample_number = 250

x_simulated = np.empty((sample_number,sample_size))
 
for i in range(sample_number):
    x_simulated[i] = vae_mft.predict(x_full_scaled[[random.randrange(x_full_scaled.shape[0])]])
```

```python
x_full = np.concatenate([x_full_scaled, x_simulated])
```

```python
print(x_full.shape)
```

```python
rev_x = scaler_vae.inverse_transform(x_full)
print(rev_x)
```

```python
x = rev_x[:, 0:-1]

y = rev_x[:, -1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Training

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor()

params_rf = {'n_estimators': [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700],
            'max_depth': [3,4,5,6,7]}

rf_gs = GridSearchCV(rf, params_rf, cv=5)
rf_gs.fit(x_train, y_train)

print(rf_gs.best_params_)
```

```python
rf = RandomForestRegressor(n_estimators = rf_gs.best_params_['n_estimators'],
                          max_depth = rf_gs.best_params_['max_depth'])

rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
```

```python
print(y_pred[::10], y_test[::10])
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(y_pred, y_test), np.sqrt(mean_squared_error(y_pred, y_test)))
print(r2_score(rf.predict(x_train), y_train), np.sqrt(mean_squared_error(rf.predict(x_train), y_train)))
```

```python
import pandas as pd

feature_imp = pd.Series(rf.feature_importances_)

print(feature_imp)
```
