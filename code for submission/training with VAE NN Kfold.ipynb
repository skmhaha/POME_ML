```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
scaler_x = StandardScaler()
x_train_scaled = scaler_x.fit_transform(x_train)
#x_valid_scaled = scaler_x.transform(x_valid)
x_test = scaler_x.transform(x_test)
```

```python
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1))
#y_valid_scaled = scaler_y.transform(y_valid.reshape(-1,1))
y_test = scaler_y.transform(y_test.reshape(-1,1))

#y_train_scaled = y_train
#y_valid_scaled = y_valid
#y_test = y_test
```

```python
def r2_score(y_true, y_pred):
    SS_res = keras.backend.sum(keras.backend.square(y_true - y_pred))
    SS_tot = keras.backend.sum(keras.backend.square(y_true - keras.backend.mean(y_true)))
    return (1 - SS_res / (SS_tot + keras.backend.epsilon()))
```

# Training

```python
import pandas as pd
import os
import numpy as np
from sklearn import metrics
from scipy.stats import zscore
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.metrics import r2_score
from tensorflow.keras import backend as K
from matplotlib import pyplot

# Cross-Validate
kf = KFold(5, shuffle=True, random_state=42) # Use for KFold classification
oos_y = []
oos_pred = []

fold = 0
for train, valid in kf.split(x_train_scaled):
    K.clear_session()
    fold+=1
    print(f"Fold #{fold}")
        
    x_train = x_train_scaled[train]
    y_train = y_train_scaled[train]
    x_valid = x_train_scaled[valid]
    y_valid = y_train_scaled[valid]
    
    if fold == 1:
        model = keras.models.Sequential([
        keras.layers.Dense(24, activation="relu", input_shape=x_train.shape[1:]),
        keras.layers.Dense(28, activation="relu"),
        keras.layers.Dense(1)])
        optimizer = tf.keras.optimizers.SGD(learning_rate=0.1,
                                        momentum=0.9,
                                        nesterov=True)


        model.compile(loss="mse", optimizer=optimizer)

        model.fit(x_train, y_train, epochs = 150, validation_data=(x_valid, y_valid), verbose=0)
        pred = model.predict(x_valid)

        oos_y.append(scaler_y.inverse_transform(y_valid))
        oos_pred.append(scaler_y.inverse_transform(pred))    

        # Measure this fold's RMSE
        print('training',
              np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train))),
             r2_score(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train)))
        
        
        score = np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(pred),
                                                   scaler_y.inverse_transform(y_valid)))
        r2 = r2_score(pred,y_valid)
        print(f"Fold score (RMSE): {score}", 'r2 =', r2)
        
                        
        r2 = r2_score(model.predict(x_test), y_test)
        print('model vs holdout :',
             'rmse', np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_test)),
                                                 scaler_y.inverse_transform(y_test))),
             'r2 =', r2)
        
        model.save('NN_Kfold_1.h5')
    
    elif fold == 2:
        model = keras.models.Sequential([
        keras.layers.Dense(112, activation="relu", input_shape=x_train.shape[1:]),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(32, activation="relu"),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(1)])


        optimizer = tf.keras.optimizers.SGD(learning_rate=0.1,
                                        momentum=0.9,
                                        nesterov=True)


        model.compile(loss="mse", optimizer=optimizer)

        model.fit(x_train, y_train, epochs = 150, validation_data=(x_valid, y_valid), verbose=0)
        pred = model.predict(x_valid)

        oos_y.append(y_valid)
        oos_pred.append(pred)    

        # Measure this fold's RMSE
        print('training',
              np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train))),
             r2_score(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train)))
        
        
        score = np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(pred),
                                                   scaler_y.inverse_transform(y_valid)))
        r2 = r2_score(pred,y_valid)
        print(f"Fold score (RMSE): {score}", 'r2 =', r2)
        
                        
        r2 = r2_score(model.predict(x_test), y_test)
        print('model vs holdout :',
             'rmse', np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_test)),
                                                 scaler_y.inverse_transform(y_test))),
             'r2 =', r2)
        
        model.save('NN_Kfold_2.h5')
    
    elif fold == 3:
        model = keras.models.Sequential([
        keras.layers.Dense(36, activation="selu", input_shape=x_train.shape[1:]),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(116, activation="relu"),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(1)])


        optimizer = tf.keras.optimizers.SGD(learning_rate=0.01,
                                    momentum=0.9,
                                    nesterov=True)


        model.compile(loss="mse", optimizer=optimizer)

        model.fit(x_train, y_train, epochs = 150, validation_data=(x_valid, y_valid), verbose=0)
        pred = model.predict(x_valid)

        oos_y.append(y_valid)
        oos_pred.append(pred)    

        # Measure this fold's RMSE
        print('training',
              np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train))),
             r2_score(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train)))
        
        
        score = np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(pred),
                                                   scaler_y.inverse_transform(y_valid)))
        r2 = r2_score(pred,y_valid)
        print(f"Fold score (RMSE): {score}", 'r2 =', r2)
        
                        
        r2 = r2_score(model.predict(x_test), y_test)
        print('model vs holdout :',
             'rmse', np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_test)),
                                                 scaler_y.inverse_transform(y_test))),
             'r2 =', r2)

        model.save('NN_Kfold_3.h5')
        
    elif fold == 4:
        model = keras.models.Sequential([
        keras.layers.Dense(12, activation="selu", input_shape=x_train.shape[1:]),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(92, activation="relu"),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(1)])


        optimizer = tf.keras.optimizers.SGD(learning_rate=0.01,
                                        momentum=0.9,
                                        nesterov=True)


        model.compile(loss="mse", optimizer=optimizer)

        model.fit(x_train, y_train, epochs = 150, validation_data=(x_valid, y_valid), verbose=0)
        pred = model.predict(x_valid)

        oos_y.append(y_valid)
        oos_pred.append(pred)    

        # Measure this fold's RMSE
        print('training',
              np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train))),
             r2_score(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train)))
        
        
        score = np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(pred),
                                                   scaler_y.inverse_transform(y_valid)))
        r2 = r2_score(pred,y_valid)
        print(f"Fold score (RMSE): {score}", 'r2 =', r2)
        
                        
        r2 = r2_score(model.predict(x_test), y_test)
        print('model vs holdout :',
             'rmse', np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_test)),
                                                 scaler_y.inverse_transform(y_test))),
             'r2 =', r2)
        
        model.save('NN_Kfold_4.h5')
    
    elif fold == 5:
        model = keras.models.Sequential([
        keras.layers.Dense(124, activation="selu", input_shape=x_train.shape[1:]),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(20, activation="sigmoid"),
        #                  kernel_regularizer=keras.regularizers.l1_l2(0.01)),
        #keras.layers.Dropout(0.2),
        keras.layers.Dense(1)])


        optimizer = tf.keras.optimizers.SGD(learning_rate=0.1,
                                        momentum=0.9,
                                        nesterov=True)


        model.compile(loss="mse", optimizer=optimizer)

        model.fit(x_train, y_train, epochs = 150, validation_data=(x_valid, y_valid), verbose=0)
        pred = model.predict(x_valid)

        oos_y.append(y_valid)
        oos_pred.append(pred)    

        # Measure this fold's RMSE
        print('training',
              np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train))),
             r2_score(scaler_y.inverse_transform(model.predict(x_train)),
                                                 scaler_y.inverse_transform(y_train)))
        
        
        score = np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(pred),
                                                   scaler_y.inverse_transform(y_valid)))
        r2 = r2_score(pred,y_valid)
        print(f"Fold score (RMSE): {score}", 'r2 =', r2)
        
                        
        r2 = r2_score(model.predict(x_test), y_test)
        print('model vs holdout :',
             'rmse', np.sqrt(metrics.mean_squared_error(scaler_y.inverse_transform(model.predict(x_test)),
                                                 scaler_y.inverse_transform(y_test))),
             'r2 =', r2)
        
        model.save('NN_Kfold_5.h5')
        

```
