```python
from sklearn.model_selection import StratifiedKFold
from scipy.stats import ks_2samp

def create_folds(df, n_s=5, n_grp=None):
    df['Fold'] = -1
    
    if n_grp is None:
        skf = KFold(n_splits=n_s)
        target = df.iloc[:,5]
    else:
        skf = StratifiedKFold(n_splits=n_s)
        df['grp'] = pd.cut(df.iloc[:,5], n_grp, labels=False)
        target = df.grp
    
    for fold_no, (t, v) in enumerate(skf.split(target, target)):
        df.loc[v, 'Fold'] = fold_no
    return df
```

```python
plt.hist(x.iloc[:,5], bins=100, density=True)
plt.xlabel('TVA')
plt.ylabel('Frequency')
plt.show()
```

```python
df = create_folds(x, n_s=4, n_grp=1000)
```

```python
for fold in np.sort(x.Fold.unique())[1:]:
    print(f'Fold 0 vs {fold}:', ks_2samp(x.loc[x.Fold==0, 5], x.loc[x.Fold==fold,5]))
```

```python
x0, y0 = None, None
x1, y1 = None, None
x2, y2 = None, None
x3, y3 = None, None

for fold in np.sort(x.Fold.unique()):

    if fold == 0:
        x0 = pd.concat([x.loc[x.Fold==1,[0,1,2,3,4]],
                        x.loc[x.Fold==2,[0,1,2,3,4]],
                        x.loc[x.Fold==3,[0,1,2,3,4]]])
        y0 = pd.concat([x.loc[x.Fold==1,5],
                        x.loc[x.Fold==2,5],
                        x.loc[x.Fold==3,5]])
    elif fold == 1:
        x1 = pd.concat([x.loc[x.Fold==0,[0,1,2,3,4]],
                        x.loc[x.Fold==2,[0,1,2,3,4]],
                        x.loc[x.Fold==3,[0,1,2,3,4]]])
        y1 = pd.concat([x.loc[x.Fold==0,5],
                        x.loc[x.Fold==2,5],
                        x.loc[x.Fold==3,5]])
    elif fold == 2:
        x2 = pd.concat([x.loc[x.Fold==0,[0,1,2,3,4]],
                        x.loc[x.Fold==1,[0,1,2,3,4]],
                        x.loc[x.Fold==3,[0,1,2,3,4]]])
        y2 = pd.concat([x.loc[x.Fold==0,5],
                        x.loc[x.Fold==1,5],
                        x.loc[x.Fold==3,5]])
    elif fold == 3:
        x3 = pd.concat([x.loc[x.Fold==0,[0,1,2,3,4]],
                        x.loc[x.Fold==1,[0,1,2,3,4]],
                        x.loc[x.Fold==2,[0,1,2,3,4]]])
        y3 = pd.concat([x.loc[x.Fold==0,5],
                        x.loc[x.Fold==1,5],
                        x.loc[x.Fold==2,5]])
```

```python
scaler_x = StandardScaler()
x0 = scaler_x.fit_transform(x0)
x1 = scaler_x.transform(x1)
x2 = scaler_x.transform(x2)
x3 = scaler_x.transform(x3)
x_test = scaler_x.transform(x_test)

scaler_y = StandardScaler()
y0 = scaler_y.fit_transform(y0)
y1 = scaler_y.transform(y1)
y2 = scaler_y.transform(y2)
y3 = scaler_y.transform(y3)
y_test = scaler_y.transform(y_test.reshape(-1,1))
```

```python
def model_builder(hp):
    hp_units_1 = hp.Int('units1', 0, 128, 4, default=8)
    hp_units_2 = hp.Int('units2', 0, 128, 4, default=8)
    activation_1 = hp.Choice('activation1', values=['sigmoid', 'relu','selu'], default='relu')
    activation_2 = hp.Choice('activation2', values=['sigmoid', 'relu','selu'], default='relu')
    
    model = keras.models.Sequential([

    
    keras.layers.Dense(units=hp_units_1, activation=activation_1, input_shape=x_train.shape[1:]),
    keras.layers.Dense(units=hp_units_2, activation=activation_2),
    keras.layers.Dense(1)])
    
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-3, 1e-2, 1e-1])

    model.compile(optimizer=keras.optimizers.SGD(learning_rate=hp_learning_rate,
                                                momentum=0.9,
                                                nesterov=True),
                loss='mse',
                metrics=['mse'])

    return model
```

```python
import pandas as pd
import os
import numpy as np
from sklearn import metrics
from scipy.stats import zscore
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.metrics import r2_score
import keras_tuner as kt
from tensorflow.keras import backend as K

# Cross-Validate

for i in range(1, 5):
    x_train = eval(f"x{i-1}")
    y_train = eval(f"y{i-1}")
     
    tuner = kt.Hyperband(model_builder,
                         kt.Objective('mse', direction="min") , max_epochs=50,
                         project_name=f'tuner_with_VAE_KStratified_{i}')

    tuner.search(x_train, y_train, epochs=50, validation_data=(x_test, y_test))


```
